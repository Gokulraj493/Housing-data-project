```python
import pandas as pd
```

```python
import os
os.getcwd()
```

```python
df=pd.read_csv("Housing-2-1.csv")
```

```python
df
```

```python
df.info()
```

```python
from sklearn.linear_model import LinearRegression ## import the necessary libraries
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
X=df[["area"]] ## segregate the data into X and Y ,2 dimentiontal
y=df[["price"]]
```

```python
X
```

```python
y
```

```python
X.isnull().sum()
```

```python
y.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split ## Spliting the data into test and train
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
model=LinearRegression()
```

```python
model
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
X_test
```

```python
import numpy as np
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
mape=np.mean(np.abs((y_test-y_pred)/y_test))*100 ## Model Evaluation
```

```python
print("r1 score is :",r2)
print("mape is :",mape)
print("mae is :",mae)
print("mse is :",mse)
```

```python
print("y intercept is",model.intercept_)
print("slops is :",model.coef_[0])
```

```python
## plot the (Training data with regressionline)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.scatter(X_train, y_train, color='black', label='Train data')
plt.plot(X_train, model.predict (X_train),color='green', label='Regression Line')
plt.scatter(X_test,y_test, color='red',label='test data')
plt.xlabel("area")
plt.ylabel('price')
plt.title('price vs. area')
plt.grid(True)
plt.show()
```
